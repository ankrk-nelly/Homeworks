```python
#  Boxplot of Lung Cancer Deaths Distribution
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\nelly\Desktop\Classes\Masters degree\Spring 2025\Data Visualization\Homeworks\Homework_#2\lung_cancer_prediction_dataset.csv')

country_deaths = df.groupby('Country')['Annual_Lung_Cancer_Deaths'].first().reset_index()

plt.figure(figsize=(12, 6))
sns.boxplot(data=country_deaths, x='Annual_Lung_Cancer_Deaths')
plt.title('Boxplot of Annual Lung Cancer Deaths by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram of PM2.5 AQI Values
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\nelly\Desktop\Classes\Masters degree\Spring 2025\Data Visualization\Homeworks\Homework_#2\global_air_pollution_dataset.csv')

print(df.columns)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(df['PM2.5_AQI_Value'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of PM2.5 AQI Values')
plt.xlabel('PM2.5 AQI Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Density Plot of the Lung Cancer Mortality Rate
mortality_rates = df[df['Mortality_Rate'] > 0]['Mortality_Rate']

plt.figure(figsize=(12, 6))
sns.kdeplot(mortality_rates, fill=True)  
plt.title('Density Plot of Lung Cancer Mortality Rate')
plt.xlabel('Mortality Rate')
plt.show()
```

```python
# Part 4: matplotlib library for this graph
import matplotlib.pyplot as plt

print(df.columns)  

plt.figure(figsize=(10, 6))

plt.hist(df['PM2.5_AQI_Value'], bins=30, density=True, color='blue', alpha=0.4, edgecolor='black', label="Histogram")
sns.kdeplot(df['PM2.5_AQI_Value'], color='darkred', linewidth=2, fill=True, alpha=0.3, label="Density Plot (KDE)")

plt.title('PM2.5 AQI Distribution with Density Overlay', fontsize=14, fontweight='bold')
plt.xlabel('PM2.5 AQI Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.figtext(0.5, -0.05, "This plot represents the distribution of PM2.5 AQI values with a density overlay.", 
            wrap=True, horizontalalignment='center', fontsize=10)
plt.show()
```
